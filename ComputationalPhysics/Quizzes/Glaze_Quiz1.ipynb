```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def Pressure(z,H=10):
    P0 = 100 #atm
    return(P0*np.exp(-z/H))
```

```python
xdata = np.linspace(0,100,100)
y10 = Pressure(xdata)
y30 = Pressure(xdata,H=30)
y100 = Pressure(xdata,H=100)

plt.scatter(xdata,y10,s=1)
plt.scatter(xdata,y30,s=1)
plt.scatter(xdata,y100,s=1)
plt.xlabel("z (km)")
plt.ylabel("P (atm)")
plt.title("Pressure vs. altitude")
plt.legend(["H = 10 km","H = 30 km","H = 100 km"])
plt.xlim(0,100)
plt.savefig("Glaze_Quiz1_Output.pdf")
```

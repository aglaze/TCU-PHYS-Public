```python
from numpy import random as rd
import numpy as np
from matplotlib import pyplot as plt
```

## Exercise 1:

```python
def e_x(x):
    return(np.sqrt(x))
```

```python
elist = []
for i in range(1000):
    elist.append(e_x(rd.rand()))
```

```python
xlist = rd.rand(1000)
elist = e_x(xlist)
```

```python
plt.hist(elist,bins=20);
plt.xlabel('sqrt(x)')
plt.ylabel('multiplicity')
plt.savefig("Glaze_Random_Fig.pdf")
```

## Exercise 2:

```python
def distance(x,y):
    return(np.sqrt(x**2 + y**2))
```

```python
dist_list = []
step_list = []

x=0
y=0
for i in range(1000000):
    prob = rd.rand()
    if prob<0.25:
        x = x+1
    elif prob<0.5:
        x = x-1
    elif prob<0.75:
        y = y+1
    else:
        y = y-1
    step_list.append(i+1)
    dist_list.append(distance(x,y))
```

```python
plt.scatter(step_list,dist_list,s=1)
plt.xlabel('number of steps, n')
plt.ylabel('distance from the origin')
plt.xscale('log')
plt.yscale('log')
```

```python
plt.scatter(step_list,dist_list,s=1)
plt.xlabel('number of steps, n')
plt.ylabel('distance from the origin')
```

### Comment on the results:
The distance from the origin due to random motion appears to have no clear trend as a function of the step number when using the linear plot. Using log scales on both axes, there appears to be a positive trend.

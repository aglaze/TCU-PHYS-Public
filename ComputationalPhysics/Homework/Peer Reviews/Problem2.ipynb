$ \Phi (x,y) = -\frac{1}{\sqrt{1+\alpha x^2 + \beta y^2}} $

```python
def RK4(func,ic1,ic2,y,h,t=0):
    k1 = h*func(ic1,y,t)
    k2 = h*func(ic1+k1/2,y,t+h/2)
    k3 = h*func(ic1+k2/2,y,t+h/2)
    k4 = h*func(ic1+k3,y,t+h)
    return ic2 + (k1+2*k2+2*k3+k4)/6

def frkx1(xd,y,t,alpha=2,beta=2): return xd
def frkx2(x,y,t,alpha=2,beta=2): 
    return -(2*beta*y)/(.5*(1+alpha*x**2+beta*y**2)**(3/2))
```

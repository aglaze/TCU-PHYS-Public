```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
'''
Created on Thu Oct 31 08:30:55 2019

@author: dcr

Example 9.3 from the textbook (heat equation IVP in 1D), with mods
'''

import matplotlib.pyplot as plt
import numpy as np
import glob
from PIL import Image
from matplotlib.animation import ArtistAnimation as animate
from IPython.display import clear_output
```

```python
# Constants
L = 0.01     # Thickness of steel [m]
D = 4.25e-6  # Thermal diffusivity [SI units]
N = 100      # Number of divisions in grid
a = L / N    # Grid spacing
h = 1.0e-4   # Time step

epsilon = h / 1000


# Output times
t1 = 0.01
t2 = 0.1
t3 = 0.4
t4 = 1.0
t5 = 10.0

tend = t5 + epsilon
```

```python
#Initial conditions
Tlo = 0.0    # Low temperature [C]
Tmid = 20.0  # Intermediate temperature [C]
Thi = 50.0   # High temperature [C]
```

```python
# Create arrays
T = np.empty(N+1, float)
T[0] = Thi
T[N] = Tlo
T[1:N] = Tmid
```

```python
# Main loop
t = 0.0
c = h * D / a**2
while t < tend:

    # Calculate the new values of T
    T[1:N] = T[1:N] + c*(T[0:N-1] - 2*T[1:N] + T[2:N+1])
    t += h

    # Make plots at the given times
    if abs(t - t1) < epsilon:  # could also use np.isclose()
        plt.plot(T, label=fr'$t$ = {t1} s')
    if abs(t - t2) < epsilon:
        plt.plot(T, label=fr'$t$ = {t2} s')
    if abs(t - t3) < epsilon:
        plt.plot(T, label=fr'$t$ = {t3} s')
    if abs(t - t4) < epsilon:
        plt.plot(T, label=fr'$t$ = {t4} s')
    if abs(t - t5) < epsilon:
        plt.plot(T, label=fr'$t$ = {t5} s')

#Output plots

plt.xlabel(r'Position $x$ [cm]')
plt.ylabel(r'Temperature $T$ [C]')
plt.legend()
plt.title('1D Heat Equation IVP')
plt.show()
```

```python
def make_animation(end_t):
    # Create arrays
    T = np.empty(N+1, float)
    T[0] = Thi
    T[N] = Tlo
    T[1:N] = Tmid
    
    h = 1e-4
    t = 0
    c = h * D / a**2
    dur = 300
    plotcount = 0
    
    while t<end_t:
        
        # Calculate the new values of T
        T[1:N] = T[1:N] + c*(T[0:N-1] - 2*T[1:N] + T[2:N+1])
        t += h


        if plotcount%100 == 0:
            print(t)
            
            clear_output(wait=True)
            #Output plots
            plt.plot(T, label=fr'$t$ = {t1} s')
            
            plt.xlabel(r'Position $x$ [cm]')
            plt.ylabel(r'Temperature $T$ [C]')
            plt.title(t)
            plt.xlim((0,100))
            plt.ylim((0,50))
            #plt.show()

            #saves the time step as a png
            filename = str("./Images/frame"+str(t).zfill(4)+".png")
            plt.savefig(filename)
            plt.clf()
        plotcount += 1
    #use the saved figures to create the frames for the animation
    frames = []
    imgs = glob.glob("./Images/*.png")
    for i in imgs:
        new_frame = Image.open(i)
        frames.append(new_frame)
    #creates an animated GIF using the created frames

    frames[0].save('1Dheat.gif', format='GIF',
                   append_images=frames[1:],
                   save_all=True,
                   duration=dur, loop=0)
#     #Output plots

#     plt.xlabel(r'Position $x$ [cm]')
#     plt.ylabel(r'Temperature $T$ [C]')
#     plt.legend()
#     plt.title('1D Heat Equation IVP')
#     plt.show()
        
```

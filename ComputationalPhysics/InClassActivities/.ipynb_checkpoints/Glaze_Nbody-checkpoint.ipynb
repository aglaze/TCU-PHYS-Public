# Section 1: Solving Kepler Equations

This section solves the Kepler Equations to establish a baseline for what the N Body code results should look like

```python
def relaxation(f,i,p):
    #arguments of f, the function, i, the initial guess, and p, the precision
    x=i
    num_iters = 0
    while(np.abs(f(x)-x)>p):
        x = f(x)
        num_iters += 1
    return(x,num_iters)
```

```python
#system variables
m1 = 1
m2 = 1
a = 1
e = 0.5

t_list = np.linspace(0,2*np.pi,100)

def psi(x,t):
    return(np.sqrt(m1+m2/(a**3))*t + e*np.sin(x))

def psidot(xd,x):
    return(np.sqrt(m1+m2/(a**3)) + e*np.cos(x)*xd)
```

```python
def relax_solve_kepler(t_list,a=a,e=e,m1=m1,m2=m2):
    
    #initializing lists
    psi_list = []
    psidot_list = []
    
    
    for t in t_list:
        #nested functions for each timestep
        def f(x):
            return(psi(x,t))

        psi_step = relaxation(f,0,0.001)[0]
        
        def fdot(xd):
            return(psidot(xd,psi_step))
        
        psidot_step = relaxation(fdot,0,0.001)[0]
        
        psi_list.append(psi_step)
        psidot_list.append(psidot_step)
        
    return(psi_list,psidot_list)
```

```python
p,pd = relax_solve_kepler(t_list,a,e,m1,m2)
```

```python
#defining position and velocity
def r(psi,a=a,e=e):
    return(a*(1-e*np.cos(psi)))

def rd(psi,psidot,a=a,e=e):
    return(a*e*np.sin(psi)*psidot)
```

```python
#constructing numpy arrays in order to get r's and rd's
p = np.array(p)
pd = np.array(pd)

r_list = r(p)
rd_list = rd(p,pd)
```

```python
plt.plot(t_list,r_list)
plt.plot(t_list,rd_list)
plt.legend(("r","rdot"))
plt.xlabel("time")
plt.xlim((0,2*np.pi/(np.sqrt(2))))
plt.title("Solution of Kepler Equations")
#plt.savefig("Glaze_2Body_Kepler.pdf")
```

```python
def GenerateIC(num_particles,rmax=5,vmax=5,mmax=5):
    #generates random ICs for specified number of particles with specified
    #    maximum position and velocities
    
    s = []
    m_list = []
    #this r_list is [x,y,z,vx,vy,vz]
    
    for i in range(num_particles):

        #generates random values for the position and velocity
        s.append(np.random.rand()*rmax)
        s.append(np.random.rand()*vmax)
        s.append(np.random.rand()*rmax)
        s.append(np.random.rand()*vmax)
        s.append(np.random.rand()*rmax)
        s.append(np.random.rand()*vmax)
        
        #and a randomized mass
        m_list.append(np.random.rand()*mmax)
        
    return(s,m_list)
```

```python
#from notes handouts:

def gravity(s, t, eps2, m):
    #inputs are
    #    state vector, s
    #    current time (not used)
    #    eps2, the softening factor squared
    #    mass array, m
    
    N = len(m) #gets number of particles
    pos = s[0::2].reshape(N, 3) #constructs positions list
    #acc = np.zeros_like(pos) #initializes acceleration list
    acc = np.zeros(np.shape(pos)) #initializes acceleration list
    
    for i in range(N - 1):
        for j in range(i + 1, N):
            r = pos[i] - pos[j] #r vector
            r2 = r@r + eps2 #calculates r^2
            a = - r / (np.sqrt(r2) * r2)
            acc[i] += m[j] * a
            acc[j] -= m[i] * a
            
    #d = np.empty(6*N) #initializes list of derivatives
    d = np.zeros(6*N) #initializes list of derivatives
    d[0::2] = s[1::2] #gets velocities from state vector
    d[1::2] = acc.flatten() #gets calculated accelerations
    #print(d)
    #print(d.dtype)
    return d


def leapfrog(s, t, h, derivs, args=()):
    #inputs of state vector, time, step size, derivative function, and 'args'
    #here, 'args' is eps2 and m
    
    s[0::2] += 0.5 * h * s[1::2] # opening drift
    s[1::2] += h * derivs(s, t + 0.5*h, *args)[1::2] # kick
    s[0::2] += 0.5 * h * s[1::2] # closing drift
    return s
```

```python
def NBody(init_s, m_list, eps, t_final):
    
    G = 1    
    N = len(m_list)
    M = np.sum(m_list)/N
    
    if eps!=0:
        #calculating timestep from average mass:
        dt = np.pi/15 * eps**(3/2) /np.sqrt(G*M)
    else:
        dt = 0.01
    
    t = 0
    s = np.copy(init_s)
    
    t_list = []
    s_list = []
    t_list.append(t)
    s_list.append(s)

    
    while t<t_final:
        
        s = np.copy(leapfrog(s,t,dt,gravity, args=(eps**2, m_list)))
        t = t+dt
        t_list.append(t)
        s_list.append(s)
        

    return(s_list,t_list)
```

```python
def gimme_stuff_to_plot(s_list):
    #reorganizes output form NBody routine to make it easier to access data for plotting
    N = int(len(s_list[1])/6)
    
    output = []
    for i in range(N):
        output.append([[],[],[],[],[],[]])
        #             ( x,vx, y,vy, z,vz)

    for t in s_list:
        for i in range(len(t)):
            p = int(i/6) #tracks number of particles
            output[p][i%6].append(t[i]) #assigns relevant vector element to the output
    #we can reference output[particle][vector element] to get list
    #     of, e.g., particle 1's list of y values using
    #     output[0][2]
    return(output)
```

```python
#2-body problem IC's:
init_2body = np.array([0.5,0,0,0.71,0,0,-0.5,0,0,-0.71,0,0])
#                     x,vx,y,vy,z,vz,x,vx,y,vy,z,vz
m_2body = np.array([1,1])
```

```python
s2body,t2body = NBody(init_2body,m_2body,0.01,5)
```

```python
new_s2body = gimme_stuff_to_plot(s2body)
```

```python
plt.plot(new_s2body[0][0],new_s2body[0][2])
plt.plot(new_s2body[1][0],new_s2body[1][2])
plt.xlim((-1,1))
plt.ylim((-1,1))
plt.axis("equal")
plt.legend(("Particle 1","Particle 2"))
plt.title('"Circular" orbit')
plt.savefig("Glaze_NBody_2particles_circularorbit.pdf")
```

```python
def calcR(x1,y1,x2,y2,vx1,vy1,vx2,vy2):
    r = []
    rdot = []
    for i in range(len(x1)):
        r.append(np.sqrt((x2[i]-x1[i])**2 + (y2[i]-y1[i])**2))
        rdot.append(np.sqrt((vx2[i]-vx1[i])**2 + (vy2[i]-vy1[i])**2))
    return(r,rdot)
```

```python
r,rd = calcR(new_s2body[0][0],new_s2body[0][2],new_s2body[1][0],new_s2body[1][2],
            new_s2body[0][1],new_s2body[0][3],new_s2body[1][1],new_s2body[1][3])
```

```python
plt.plot(t2body,r)
plt.plot(t2body,rd)
plt.legend(("r","rdot"))
plt.title("Relative Positions and Velocities")
plt.savefig("Glaze_NBody_2particles_circularorbit_separations.pdf")
```

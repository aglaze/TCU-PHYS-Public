```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def phi(r,m1=3,m2=2,d=1):
    m = m1 + m2
    omegas = m/d**3
    x1 = -(m2/m)*d
    x2 = (m1/m)*d
    r1 = np.array([x1,0])
    r2 = np.array([x2,0])
    r1s = r-r1
    r2s = r-r1
    rm1 = np.sqrt(np.dot(r1s,r1s))
    rm2 = np.sqrt(np.dot(r2s,r2s))
    
    return -(m1/rm1) - (m2/rm2) - .5*omegas*np.dot(r,r)
```

```python
a = .6
b = 1
N = 100000
h = (b-a)/N
cphi = 1
x=0
y=0
pt = np.arange(a,b+h,h)
dxs = []

for i in pt:
    r = np.array([i,y])
    rh = np.array([i+h,y])
    dx = (phi(rh)-phi(r))/h
    dxs.append(dx)
    #x += h
    #y+=h
    #print(cphi,x)
    if abs(dx) < 1e-6:
        break
```

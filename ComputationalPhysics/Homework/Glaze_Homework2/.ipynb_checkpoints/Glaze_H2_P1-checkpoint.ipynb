```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Analytical derivative of e^x is e^x...
#    evaluated at x = 0, d/dx(e^x) = 1

def frdDiff(f,a,b,h,N):
    #forward difference method
    xvals = np.linspace(a,b,N)
    return((f(xvals+h)-f(xvals))/h)

def cenDiff(f,a,b,h,N):
    #central difference method
    xvals = np.linspace(a,b,N)
    return((f(xvals+h/2)-f(xvals-h/2))/h)

def ddxE_forward(h):
    #takes derivative of e^x and evaluates at x = 0
    #calls forward difference method function with a given step size, h
    
    func = frdDiff(np.exp,-1,1,h,101)
    #min & max range, as well as number of points is hard-coded, but are arbitrary...
    
    return(func[50]) #returns the function at the index corresponding to x = 0

def ddxE_center(h):
    #takes derivative of e^x and evaluates at x = 0
    #calls forward difference method function with a given step size, h
    
    func = cenDiff(np.exp,-1,1,h,101)
    #min & max range, as well as number of points is hard-coded, but are arbitrary...
    
    return(func[50]) #returns the function at the index corresponding to x = 0
```

## Using the forward difference method:

```python
hlist = np.logspace(-16,0,100) #100-item list in log space from 1e-16 to 1


#currently, won't work with normal numpy list methods (i.e., can't define ddxE(hlist) due to shape issues)
#so I use the ugly but tried-and-true method:
absdiff_fwd = []
for h in hlist:
    absdiff_fwd.append(np.abs(ddxE_forward(h)-1))

#and plot...
plt.scatter(hlist,absdiff_fwd)
plt.xscale('log')
plt.xlabel("h")
plt.ylabel("absolute difference")
plt.title("Difference between Analytical and Numerical derivative vs step size");
```

```python
#not sure if lin-log or log-log axes are best....
#    you decide?

plt.scatter(hlist,absdiff_fwd)
plt.xscale('log')
plt.yscale('log')
plt.xlabel("h")
plt.ylabel("absolute difference")
plt.title("Difference between Analytical and Numerical derivative vs step size");
```

## Using the center difference method:

```python
absdiff_ctr = []
for h in hlist:
    absdiff_ctr.append(np.abs(ddxE_center(h)-1))

#and plot...
plt.scatter(hlist,absdiff_ctr)
plt.xscale('log')
plt.xlabel("h")
plt.ylabel("absolute difference")
plt.title("Difference between Analytical and Numerical derivative vs step size");
```

```python
#not sure if lin-log or log-log axes are best....
#    you decide?

plt.scatter(hlist,absdiff_ctr)
plt.xscale('log')
plt.yscale('log')
plt.xlabel("h")
plt.ylabel("absolute difference")
plt.title("Difference between Analytical and Numerical derivative vs step size");
```

Both methods give good results with very small differences. However, the center difference method gives better results even at larger step sizes. See comparison below.

```python
#Comparison:
plt.scatter(hlist,absdiff_fwd)
plt.scatter(hlist,absdiff_ctr)
plt.xscale('log')
plt.yscale('log')
plt.xlabel("h")
plt.ylabel("absolute difference")
plt.legend(["Forward","Center"])
plt.title("Comparison of Forward Difference and Center Difference");
```

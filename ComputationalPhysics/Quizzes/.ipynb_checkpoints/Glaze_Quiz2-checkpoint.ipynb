```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def rho(x,r0 = 1):
    if (np.abs(x)<0.5):
        return(r0*np.cosh(x))
    else:
        return(r0*np.cos(x))

#trapezoid integrator:
def trap_int(a,b,N,f):
    #note diff between h and H:
    #    H is the thickness of the slices
    #    h is Planck's constant
    H = (b-a)/N
    
    #defining the sum used in the trapezoid rule:
    trapsum = 0
    for i in range(1,N):
        trapsum += f(a+i*H)
    
    #returning the resulting area
    return(H*(1/2 * f(a) + 1/2 * f(b) + trapsum))
```

```python
M = trap_int(-1,1,100,rho)
print(M)
```

```python
M = round(M,2)
print(M)
```

```python
x_list = np.linspace(-1,1,100)
y_list = []
for i in x_list:
    y_list.append(rho(i))

plt.scatter(x_list,y_list,cmap='copper',c = y_list)
plt.xlim((-1,1))
plt.ylim((0,2))
plt.title(str("Density as a function of position; Total mass = " + str(M)))
plt.xlabel("x")
plt.ylabel("rho(x)")
plt.savefig("Glaze_Quiz2_Output.pdf")
```

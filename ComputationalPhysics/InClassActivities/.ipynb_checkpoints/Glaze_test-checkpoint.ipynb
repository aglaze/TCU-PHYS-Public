```python
#standard practice is to import all necessary packages in the first cell of the notebook
import numpy as np
from matplotlib import pyplot as plt
import scipy
```

```python
#then we define the functions that we will use later
#example: function that takes in an integer, and outputs that value squared
def square_that_integer(input_int):
       return(input_int*input_int)
```

```python
#now we want to create an array with 100 values ranging from 0 to 10
#then we run our squaring function on those values, and plot the results as a scatter plot
x_values = np.linspace(0,1,num=100)
y_values = square_that_integer(x_values)
```

```python
plt.scatter(x_values,y_values,s=2)
```

```python
#2 sets of 100 random numbers between 0 and 1
rand_list1 = np.random.normal(0.5,0.2,100)
rand_list2 = np.random.normal(0.5,0.2,100)
colors = rand_list1*rand_list2
```

```python
x_values = np.linspace(0,1,num=100)
plt.scatter(rand_list1,rand_list2,s=10/(colors),alpha=0.7,c=colors,cmap="gist_stern_r")
```
